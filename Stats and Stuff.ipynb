## Create Null distributions
- 14000 cells
- Assume the frequency of cell types is independent from the healthy/myeloma distinction.
- 10000

```python
cell_abundances = [0.132,
                   0.237,
                   0.213,
                   0.088,
                   0.114,
                   0.208,
                   0.009] ## according to H_0

n_cells = 14029
n_trials = 10000

null_a,null_b,null_c,null_d,null_e,null_f,null_g  = [],[],[],[],[],[],[]

for i in tqdm(range(n_trials)):
    
    type_a = 0
    type_b = 0
    type_c = 0
    type_d = 0
    type_e = 0
    type_f = 0
    type_g = 0
    
    for j in range(n_cells):
        rand_value = np.random.rand()
        
        if rand_value < 0.132:
            type_a += 1
            
        elif rand_value > 0.132 and rand_value < (0.132 + 0.237):
            type_b += 1
            
        elif rand_value > (0.132 + 0.237) and rand_value < (0.132 + 0.237 + 0.213):
            type_c += 1
            
        elif rand_value > (0.132 + 0.237 + 0.213) and rand_value < (0.132 + 0.237 + 0.213 + 0.088):
            type_d += 1
        
        elif rand_value > (0.132 + 0.237 + 0.213 + 0.088) and rand_value < (0.132 + 0.237 + 0.213 + 0.088 + 0.114):
            type_e += 1
        
        elif rand_value > (0.132 + 0.237 + 0.213 + 0.088 + 0.114) and rand_value < (0.132 + 0.237 + 0.213 + 0.088 + 0.114 + 0.208):
            type_f += 1
        else:
            type_g += 1
            
    null_a.append(type_a/n_cells)
    null_b.append(type_b/n_cells)
    null_c.append(type_c/n_cells)
    null_d.append(type_d/n_cells)
    null_e.append(type_e/n_cells)  
    null_f.append(type_f/n_cells)            
    null_g.append(type_g/n_cells)            

        
print("DONE")
```

```python
null_dist_deltas_list = []

for i in range(len(cell_abundances)):
    n_healthy_a = 6000*healthy_ras[i]
    n_diseased_a = 8000*diseased_ras[i]

    n_bootstraps = 1000

    null_dist_of_deltas = []

    observed_delta_a = healthy_ras[i]-diseased_ras[i]
    print("Observed Delta = ",abs(observed_delta_a))

    for i in tqdm(range(n_bootstraps)):

        ## Sample from null hypo the appropriate number of times (aka the number of cell type in each condition)
        healthy_a_under_null = choices(null_a,k=int(n_healthy_a))
        diseased_a_under_null = choices(null_a,k=int(n_diseased_a))

        ## Calculate the mean relative abundance from the sampled distributions (these are the bars)
        mean_healthy_a = np.mean(healthy_a_under_null)
        mean_diseased_a = np.mean(diseased_a_under_null)

        ## Calculate delta
        delta_a = abs(mean_diseased_a-mean_healthy_a)

        ## Append it to the null dist list
        null_dist_of_deltas.append(delta_a)
    sns.histplot(null_dist_of_deltas)
    #plt.axvline(observed_delta_a,color='black',linestyle='--',alpha=0.5,label=r'Observed $\Delta$ A')
    plt.show()
```
